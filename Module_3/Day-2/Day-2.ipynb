```python
import pickle

student_data = {
    'name': 'Sanjai Kannan G',
    'email': 'sanjai@gmail.com',
    'class': 'XII',
    'marks': {
        'maths': 90,
        'science': 85,
        'english': 80
    }
}

# Serialize the data
result = pickle.dumps(student_data)
print(result)

with open ('student_data.pkl', 'ab') as file:
    pickle.dump(student_data, file)
    with open ('student_data.pkl', 'rb') as file:
        data = pickle.load(file)
        print(data)
```

```python
import json

student_data = {
    'name': 'Sanjai Kannan G',
    'email': 'sanjai@gmail.com',
    'class': 'XII',
    'marks': {
        'maths': 90,
        'science': 85,
        'english': 80
    }
}

# Save data in JSON format
with open('student_data.json', 'w') as file:
    json.dump(student_data, file)

# Read the JSON file
with open('student_data.json', 'r') as file:
    data = json.load(file)

# Extract the data
result = data
print(result)
```
