```python
# Grouped Bar Plot

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

course = np.arange(4)

# Male 
students =[100, 110, 120, 130]
plt.bar(course-0.2, students, width=0.4, color='r')

# Female 
students = [30, 150, 190, 70]
plt.bar(course+0.2, students, width=0.4, color='b')

plt.xticks(course, ['Python', 'Java', 'C++', 'JavaScript'])

plt.show()
```

```python
# Line Plot

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

course = np.arange(4)

# Male 
male =[100, 110, 120, 130]
plt.stem(course, male)

plt.show()
```

```python
# Stair Plot

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

course = np.arange(10)

# Male 
male =[100, 110, 120, 130, 140, 150, 160, 170, 180, 190]
plt.stairs(male)

plt.show()
```

```python
# errorbar

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

course = np.arange(10)
male =[100, 110, 10, 130, 140, 50, 60, 70, 80, 90]
error = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]

plt.errorbar(course, male, error, fmt='o', color='r')

plt.show()
```
